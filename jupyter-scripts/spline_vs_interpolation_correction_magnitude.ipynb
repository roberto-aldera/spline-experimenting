```python
import csv
import pandas as pd
import numpy as np
# To plot pretty figures
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
# plt.rc("font", family=settings.FONT_FAMILY)
plt.rcParams['axes.titlepad'] = 20
plt.rcParams.update({'font.size': 16}) # was 28
plt.rcParams['axes.grid'] = True
plt.rcParams['text.usetex'] = True

interpolation_corrections = []
spline_corrections = []

interpolation_x = []
spline_x = []
interpolation_x = []
```

```python
filename_interpolation = '/Users/roberto/data/spline-experiments/correction_matrices/interpolation/1503053637231673_G_correction_mat.csv'
filename_splines = '/Users/roberto/data/spline-experiments/correction_matrices/splines/1503053637231673_G_correction_mat.csv'

df_interpolation=pd.read_csv(filename_interpolation, sep=',',header=None)
df_splines=pd.read_csv(filename_splines, sep=',',header=None)
interpolation_x.append(df_interpolation.values[0])
spline_x.append(df_splines.values[0])

plt.figure(figsize=(20,10))
plt.plot(df_interpolation.values[0],'k.',label='Interpolation');
plt.plot(df_splines.values[0],'b^',label='Spline');
plt.legend();
print("Interpolation initial correction:",df_interpolation.values[0][0])
print("Spline initial correction:",df_splines.values[0][0])

spline_corrections.append(df_splines.values[0][0])
interpolation_corrections.append(df_interpolation.values[0][0])
```
